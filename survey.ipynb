```python
import pandas as pd 
from matplotlib import pyplot as plt
pd.set_option('display.width', 5000) 
pd.set_option('display.max_columns', 60) 
```

## read data in

```python
df = pd.read_csv('survey.csv')

df
```

## put column names into a list

```python
df.columns.values.tolist()
```

## delete columns

```python
del df['Timestamp']
del df['plural endings of nouns end in “i“, no matter what the gender. (siciliano)']
del df['Come chiami la Farmacia?']
del df["Quando parli di un'altra persona, aggiungi la preposizione davanti a suo nome (ex: la Chiara ieri mi ha chiamato)"]
del df ['Come chiami il formaggio?']
```

```python
df
```

## rename columns

```python
df.rename(columns={"A quale regione d'Italia appartieni?": 'regione'}, inplace=True)
df.rename(columns={'Come riferisci ad un gruppo di ragazzi nel tuo dialetto?': 'ragazzi', 'Come diresti la frase ‘Ho capito’?    ': 'ho_capito'}, inplace=True)
df.rename(columns={'Come chiameresti una sedia nel tuo dialetto?': 'sedia', 'Come chiameresti una sigaretta?': 'sigaretta', 'Come chiameresti un tiro di una sigaretta?': 'tiro', 'Come pronunci la parola aspetta?': 'aspetta'}, inplace=True)
df.rename(columns={'Come dici la parola ‘Gatto’ in dialetto?' : 'gatto',
 'Come dici la parola ‘Cane’??' : 'cane'}, inplace=True)
df.rename(columns={'Come si chiama il salame dalle tue parti?' : 'salame',
 'Come chiameresti una ‘Mucca’ nel tuo regione d’Italia?' : 'mucca', 'Come chiameresti alla spezia ‘Prezzemolo’ nel tuo dialetto?' : 'prezzemolo'}, inplace=True)
```

## get counts for each response for a column (all regions)

```python
gatto_counts = df['gatto'].value_counts()
print(gatto_counts)
```

```python
%matplotlib inline
gatto_counts.plot(kind='bar', figsize=(15,10))
```

```python
ragazzi_counts = df['ragazzi'].value_counts()
ragazzi_counts.plot(kind='bar', figsize=(15,10))
```

## counts for a specific region's responses to a specific question

```python
campania_cane = df[df['regione'].str.match('Campania')]['cane']
campania_cane_counts = campania_cane.value_counts()
campania_cane_counts
campania_cane_counts.plot(kind='bar', figsize=(10,5))
```
